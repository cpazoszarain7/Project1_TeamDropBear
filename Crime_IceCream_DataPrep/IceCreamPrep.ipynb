```python
#AtlCrime2016:
path = os.path.join('..','Resources','AtlCrime2016.csv')
df = pd.read_csv(path)
df.drop('Unnamed: 0', axis=1, inplace=True)

#AtlCrime2016:-Split Date column into day, month, and year to be able to call Visual Crossing API- Rabia
dsplit = df['date'].apply(lambda x: x.split("/"))
months = []
days = []
years = []

for item in dsplit:
    months.append(item[0])
    days.append(item[1])
    years.append(item[2])

#Create Month, Day and Year columns
df['Month'] = months
df['Day'] = days
df['Year'] = years

#AtlCrime2016
crime = df.loc[df['Year']=='2016']
crime.reset_index(drop =True, inplace=True)
crime.head()
```

# Section 2: Create DataFrames for Data Wrangling

```python
#1. LARCENY-FROM VEHICLE
#2. BURGLARY-RESIDENCE
#3. ROBBERY-PEDESTRIAN
#4. RAPE
#5. HOMICIDE
cdf2= cdf.loc[(cdf["crime"]=="LARCENY-FROM VEHICLE")| (cdf["crime"]=="BURGLARY-RESIDENCE")|(cdf["crime"]=="ROBBERY-PEDESTRIAN")|(cdf["crime"]=="RAPE")|(cdf["crime"]=="HOMICIDE")]
```

```python
#Trim to two decimal places 
cdf2['lat']= round(cdf2['lat'],2)
cdf2['long']= round(cdf2['long'],2)
```

```python
#Preview the new data frame, focusing on all the crimes we selected
cdf2
```

```python
#Concatenate and Convert the latitutude/longitude columns into a single column for aggregation
cdf2['latLong']= cdf2['lat'].astype(str) + "," + cdf2['long'].astype(str)
```

```python
#Prevew newly created latlong column
cdf2
```

```python
#Create data frame to hold the unique long/lat values
cityLL =pd.DataFrame(cdf2['latLong'].unique())
```

```python
cityLL
```

# Section 3 : API Request 

```python

base_url = "https://maps.googleapis.com/maps/api/place/textsearch/json?"
radius = "1609.34" # 1 mile proximity converted to meters
max_price = "2" # maximum price set to level 2

    #location = 

params = {
"radius": radius,
"query": ["ice cream shop"],
"type": ['cafe','restaurant','drugstore'],
"maxprice": max_price,
"key": gkey,
}

for i, row in cityLL.iterrows():
    
    params["location"] = row[0]
    response = requests.get(base_url, params=params).json()
    try:
        cityLL.loc[i,'store count']= len(response['results'])
        #print(len(response['results']))
        #print(row[0])
    except:
        cityLL.loc[i,'store count']= np.nan 
        
                                
       
    
```

```python
#Data frame created to tie the coordinates to the store count based on 1 mile proximity
cityLL
```

```python
cityLL.rename(columns={0: "latLong"},inplace=True)
```

```python
#merging the total crime dataframe on latLong column
cdf3 = cdf2.merge(cityLL, on='latLong')
```
